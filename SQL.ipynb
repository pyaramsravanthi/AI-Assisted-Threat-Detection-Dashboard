## SQL Prompts and Examples

```python
import sqlite3
import pandas as pd

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create a sample table: Employees
cursor.execute('''
CREATE TABLE employees (
    id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    department TEXT NOT NULL,
    salary INTEGER NOT NULL,
    hire_date DATE
);
''')

# Insert sample data
employees_data = [
    (1, 'Alice', 'Smith', 'HR', 60000, '2020-01-15'),
    (2, 'Bob', 'Johnson', 'Engineering', 85000, '2019-03-20'),
    (3, 'Charlie', 'Brown', 'HR', 62000, '2021-06-01'),
    (4, 'Diana', 'Prince', 'Marketing', 70000, '2018-11-10'),
    (5, 'Eve', 'Davis', 'Engineering', 90000, '2020-07-22'),
    (6, 'Frank', 'White', 'Sales', 75000, '2022-01-05'),
    (7, 'Grace', 'Black', 'Marketing', 72000, '2019-09-01'),
    (8, 'Heidi', 'Green', 'Sales', 78000, '2021-04-12'),
    (9, 'Ivan', 'Blue', 'Engineering', 95000, '2018-05-30'),
    (10, 'Judy', 'Red', 'HR', 61000, '2022-03-01')
]

cursor.executemany('INSERT INTO employees VALUES (?, ?, ?, ?, ?, ?)', employees_data)
conn.commit()

print("Sample 'employees' table created and populated successfully.")
```

### Prompt 1: Select all employees

```python
# Query: Retrieve all columns and rows from the 'employees' table.
sql_query = "SELECT * FROM employees;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 2: Select employees from a specific department

```python
# Query: Get the first name, last name, and salary of all employees in the 'Engineering' department.
sql_query = "SELECT first_name, last_name, salary FROM employees WHERE department = 'Engineering';"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 3: Order employees by salary

```python
# Query: List all employees, ordered by their salary in descending order.
sql_query = "SELECT * FROM employees ORDER BY salary DESC;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 4: Count employees per department

```python
# Query: Find out how many employees are in each department.
sql_query = "SELECT department, COUNT(id) AS num_employees FROM employees GROUP BY department;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 5: Calculate average salary per department

```python
# Query: Calculate the average salary for each department.
sql_query = "SELECT department, AVG(salary) AS average_salary FROM employees GROUP BY department;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 6: Find employees with salary above a certain amount

```python
# Query: Get employees whose salary is greater than 70000.
sql_query = "SELECT first_name, last_name, salary FROM employees WHERE salary > 70000 ORDER BY salary DESC;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 7: Find employees hired after a specific date

```python
# Query: Retrieve employees who were hired after '2020-01-01'.
sql_query = "SELECT first_name, last_name, hire_date FROM employees WHERE hire_date > '2020-01-01' ORDER BY hire_date ASC;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 8: Find the highest salary in each department

```python
# Query: Determine the maximum salary within each department.
sql_query = "SELECT department, MAX(salary) AS max_salary FROM employees GROUP BY department;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 9: Employees whose names start with a specific letter

```python
# Query: List all employees whose first name starts with 'A'.
sql_query = "SELECT first_name, last_name FROM employees WHERE first_name LIKE 'A%';"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

### Prompt 10: Employees with salary within a range

```python
# Query: Find employees with a salary between 60000 and 70000 (inclusive).
sql_query = "SELECT first_name, last_name, salary FROM employees WHERE salary BETWEEN 60000 AND 70000 ORDER BY salary;"

df = pd.read_sql_query(sql_query, conn)
display(df)

# Close the database connection
conn.close()
```
